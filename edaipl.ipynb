```python
import scipy.stats as st
```

```python
def chi_square_test(table,column1,column2):
    p_value=st.chi2_contingency(table)[1] # here I am taking 1 index value because that value is p value which is only usefull further.
    cl = 0.95
    alpha =1-cl
    if p_value<=alpha:
        print("reject null hyphotesis,both are dependence")
        print(f"There is a significant association between {column1} and {column2}")
    else:
        print("failed to reject null hyphotesis,both are independence")
        print(f"There is no significant association between {column1} and {column2}")
```
